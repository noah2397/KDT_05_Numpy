```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
#22.12.31 년 기준 지하철 이용 수
pd.set_option('display.precision',3)
df=pd.read_csv("subway.csv", usecols=[2,4,5])
```

```python
station=dict()
for i in df["승차_역"].unique():
    station.setdefault(i,0)
```

```python
for i in range(238567):
    station[df.iloc[i,0]]+=df.iloc[i,2]
    station[df.iloc[i,1]]+=df.iloc[i,2]
```

```python
df2=pd.Series(station)
```

```python
#1 각 지하철 역별 평균 이용객 수 
df2
```

```python
#2 각 지하철 역별 이용객수의 표준 편차 
np.std(df2.values)
```

```python
#3 가장 많은 이용객이 있는 역의 이용객 수
df2.max()
```

```python
#3 가장 적은 이용객이 있는 역의 이용객 수
df2.min()
```

```python
#5 이용객 수가 가장 많은 역의 이름
df2.idxmax()
```

```python
#6 이용객 수가 가장 적은 역의 이름
df2.idxmin()
```
